## 📦 Setup & Prerequisites
First, let's write our updated `coach_toolbox_v2.py` helper library to disk so our dashboard can import its analytical functions cleanly.

````python

# =====================================================================
# 🧱                  THE SOCCER ANALYTICS LIBRARY --- VERSION 2.0
# =====================================================================

def calculate_goals(player_profile):
    return sum(player_profile.get("goals", []))

def calculate_assists(player_profile):
    return sum(player_profile.get("assists", []))

def calculate_historical_contributions(player_profile):
    return calculate_goals(player_profile) + calculate_assists(player_profile)

def passing_accuracy(player):
    total_comp = sum(player.get("passes_completed", []))
    total_att = sum(player.get("total_passes", []))
    if total_att == 0:
        return 0.0
    return (total_comp / total_att) * 100

def needs_passing_practice(player):
    return passing_accuracy(player) < 80.0

def needs_rest(player):
    return calculate_goals(player) > 5

def is_forward(player):
    return player.get("position") in ["LW", "CF", "RW", "ST"]

def needs_endurance_training(player):
    return is_forward(player) and passing_accuracy(player) < 85

def excellent_stamina(player):
    for cp in player.get("passes_completed", []):
        if cp <= 20:
            return False
    return True

def calculate_overall_rating(player):
    if player.get("position") == "GK":
        return (player["stamina"] * 0.40) + (player["speed"] * 0.20)
    else:
        return (player["defending"] + player["stamina"] + player["speed"] + player["shooting"]) / 4

def coach_decision(player):
    if needs_rest(player):
        return "Recovery Day"
    if needs_endurance_training(player):
        return "Endurance Training"
    if needs_passing_practice(player):
        return "Passing Practice"
    return "Ready for Match"

def generate_ai_tactical_recommendation(player):
    if player["defending"] > 80:
        return "Solid defensive anchor. Hold the backline."
    if player["speed"] > 90 and player["shooting"] > 80:
        return "Winger breakout threat. Focus on flank overlaps."
    elif player["shooting"] > 90:
        return "Target striker. Look for direct shots inside the box."
    else:
        return "Tactical supporting playmaker. Distribute cleanly."
````
