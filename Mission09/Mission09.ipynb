## 📖 Python → Streamlit Translation Guide

Use this reference table whenever you want to convert Python terminal code into web app components:

| What You Know in Python | Streamlit Equivalent | Description |
| :--- | :--- | :--- |
| `print("Hello")` | `st.write("Hello")` / `st.text("Hello")` | Displays general text or variables |
| `print("Title")` | `st.title("Title")` / `st.header("Header")` | Displays large headers |
| `input("Name: ")` | `st.text_input("Name: ")` | Text entry field |
| `int(input("Age: "))` | `st.number_input("Age: ")` | Numerical counter input |
| `input("Position: ")` | `st.selectbox("Position:", ["Forward", ...])` | Dropdown selection menu |
| `bool_val = True` | `st.checkbox("Enable Option")` | On/Off toggle switch |
| `choice = "A"` | `st.radio("Select Tactics:", ["A", "B"])` | Radio button selector |
| `display(df)` | `st.dataframe(df)` / `st.table(df)` | Interactive data tables |
| `plt.bar()` | `st.bar_chart(df)` | Renders interactive bar charts |
| `print("Error!")` | `st.error("Error!")` | Red alert message box |
| `print("Success!")` | `st.success("Success!")` | Green success alert box |
| `print("Warning!")` | `st.warning("Warning!")` | Yellow warning alert box |
| `print("Info!")` | `st.info("Info!")` | Blue informational alert box |

## 🧠 How Streamlit Thinks

In a standard Jupyter Notebook, code executes sequentially from top to bottom cell by cell:
```text
Notebook: Run Cell 1 ➔ Run Cell 2 ➔ Run Cell 3 (State updates manually)
```

**Streamlit uses a reactive model.** Every time a user clicks a button, moves a slider, or changes a dropdown, Streamlit reruns your Python script from line 1 to the end:

```text
[User Interacts with Widget]
          │
          ▼
┌─────────────────────────────────────────┐
│ Streamlit executes app.py from line 1   │
│ Reads current widget states             │
│ Re-evaluates all Python logic           │
│ Repaints the Web Dashboard in browser   │
└─────────────────────────────────────────┘
```

> 💡 **Key Mental Model:** You just write simple, top-to-bottom Python logic. Streamlit handles redrawing the screen automatically!

> *Why display static text when you can use structured UI elements like `st.metric()`? Metrics give coaches instant visual feedback on key performance indicators (KPIs).*

````python
import streamlit as st

# Player Variables
player_name = "Artin"
player_position = "Forward"
player_stamina = 90
player_speed = 92

# Streamlit Translation
st.title(f"⚽ Player Profile: {player_name}")
st.write(f"**Position:** {player_position}")

# Interactive KPIs
col1, col2 = st.columns(2)
col1.metric(label="Stamina Rating", value=f"{player_stamina}%", delta="+2%")
col2.metric(label="Speed Rating", value=f"{player_speed}/100", delta="+5%")
````
