## 🛠️ Part 1 — Setting Up Your Development Environment

Follow these step-by-step installation instructions on your physical desktop machine to configure your application framework.

### 1. Installing Python on Your Machine
* **Mac:** macOS comes with basic python resources, but you should install the complete community release package from python.org or deploy it via Homebrew using `brew install python` in the terminal app.

* **Windows:** Download the latest stable Python 3 release installer from the official python.org site. Here is the straightforward step-by-step process to install Python 3 on Windows:

**Step 1: Download the Python Installer**
Go to the official Python download page for Windows at python.org/downloads/windows.
Click the button to download the latest Python 3.x.x installer (e.g., Python 3.12 or Python 3.13).

**Step 2: Run the Installer**
Open the downloaded .exe file (usually located in your Downloads folder).
Crucial Step: At the bottom of the installation window, check the box that says:
 Add python.exe to PATH (If you miss this step, Windows won't recognize Python commands in Command Prompt or PowerShell).
Click Install Now.

**Step 3: Verify the Installation**
Open the Command Prompt or PowerShell:
Press Win + R, type cmd, and press Enter.
Run the following command to check the Python version:

`python --version`

Test pip (Python's package manager) to make sure it was installed correctly:

`pip --version`

### 2. Installing Visual Studio Code (VS Code)
VS Code is an IDE. An **Integrated Development Environment (IDE)** is the professional software developer's ultimate workspace. While Jupyter Notebooks combine documentation with code cells, an IDE is built specifically for scaling out production source code files. It provides built-in syntax highlighting, automated code completion hints, background debugging tools, and absolute management over your system's folder hierarchies. Today, we step into the exact same workflow used by senior AI engineers worldwide.
* Download and install **VS Code** from `code.visualstudio.com`.
* Once open, click on the **Extensions** marketplace icon on the far left panel (looks like 4 small blocks arranged as a square).
* Search for **Python** (by Microsoft) and click **Install** to inject syntax auto-complete capabilities directly into your editor.

### 4. Deploying Streamlit via the Terminal
* Open your terminal directly inside VS Code by pressing **`Ctrl + \``** (backtick) or selecting **Terminal -> New Terminal** from the upper navigation.
* Execute the following library package installation command:
  ```bash
  pip install streamlit
  ```
* Verify the setup is complete by launching Streamlit's built-in demo environment:
  ```bash
  streamlit hello
  ```
* *Note: If a web page opens showing the Streamlit welcome dashboard, close the terminal server by pressing `Ctrl + C` in your command line window.*

* Install the other required libraries:
```bash
  pip install pandas matplotlib seaborn
  ```

### 5. Creating your Source Code File
* In the left explorer tree panel of VS Code, click the **New File** icon and create a file named exactly **`app.py`**.

### 6. Running Your Application Code
* Whenever you want to launch your file, run this command in your open terminal panel:
  ```bash
  streamlit run app.py
  ```
* This will automatically open a local server port inside your web browser at `http://localhost:8501`.

#### 🛑 Common Beginner Mistakes to Avoid
1. **Running `streamlit run app.py` inside the wrong system folder:** Always make sure your terminal path shows you are standing inside the exact folder where `app.py` lives. Type `pwd` (Mac) or `dir` (Windows) to verify.
2. **Trying to run regular Python syntax commands while a Streamlit server is spinning:** When the server is live, the terminal is locked holding up the website. To run commands, open a second terminal tab using the `+` icon in VS Code.
3. **Forgetting to Save Files:** Streamlit cannot see code updates unless you physically save your file in VS Code (`Ctrl + S` or `Cmd + S`). Turn on **File -> Auto Save** in VS Code to fix this permanently!

## 🔹 Phase 3: Add Coach Controls 

Give Coach Artin control over the squad using interactive widgets!

### Essential Widgets:
* `st.selectbox()` - Choose items from a dropdown list
* `st.slider()` - Adjust numerical values (e.g., Stamina thresholds)
* `st.button()` - Execute tactical actions on click
* `st.checkbox()` - Toggle options on/off
* `st.radio()` - Pick single strategy options


````python
import streamlit as st

st.title("🎮 Tactical Widget Testing Lab")

# 1. Buttons
if st.button("📢 Shout Team Motivation"):
    st.success("🔥 'Let's win this match together! No excuses, maximum focus!'")

# 2. Text Input
custom_note = st.text_input("✍️ Add a private scout note:")
if custom_note:
    st.info(f"Saved note: {custom_note}")

# 3. Sliders
stamina_cutoff = st.slider("🫁 Minimum Stamina Threshold Required:", 0, 100, 85)
st.write(f"Filtering squad for players with stamina higher than: **{stamina_cutoff}**")
````

````python
import streamlit as st
import pandas as pd

st.title("⚽ AI Soccer Coach")

df = pd.read_csv("Artin_FC.csv")
df.columns = df.columns.str.strip()

st.header("🎛️ Coach Controls")

# Widget Controls
selected_pos = st.selectbox("Choose position to inspect:", df["position"].unique())
min_stamina = st.slider("Minimum Stamina:", 50, 100, 70)

# Filtering DataFrame based on widgets
filtered_df = df[(df["position"] == selected_pos) & (df["stamina"] >= min_stamina)]

st.subheader(f"Filtered Roster ({selected_pos})")
st.dataframe(filtered_df)

if st.button("Confirm Filter"):
    st.write(f"Found {len(filtered_df)} players matching criteria!")
````

## 🔹 Phase 4: Connect AI Logic

Now Streamlit becomes the visual body for decision logic. We feed input values from widgets into Python rules and output visual alerts using `st.success()`, `st.warning()`, and `st.error()`.


````python
import streamlit as st

st.title("🧠 AI Fitness & Readiness Rule Engine")

# Inputs from UI
stamina = st.slider("Player Stamina Level:", 30, 100, 65)
speed = st.slider("Player Speed Rating:", 30, 100, 80)

# Connected Decision Logic
if stamina < 50:
    advice = "⚠️ Player needs rest! High risk of fatigue or muscle injury."
    st.warning(advice)
elif speed > 85:
    advice = "🚀 High-speed threat detected! Ready to start on the wing."
    st.success(advice)
else:
    advice = "✅ Player is fit and ready for match inclusion."
    st.info(advice)
````

## 🔹 Phase 5: Build the Soccer AI Dashboard

Let's wire everything together into our final project architecture!

### Final Directory Layout:
```text
soccer_ai_coach/
│
├── app.py                 ← Main Streamlit dashboard
├── coach_toolbox_v2.py       ← Your custom helper functions
├── Artin_FC.csv           ← Player stats
```

In `app.py`, we import functions from `coach_toolbox.py`, read `Artin_FC.csv`, arrange layout elements into columns (`st.columns`), display metrics, and visualize player stats using `st.bar_chart()`.


````python
import streamlit as st
import pandas as pd

# Importing your custom toolbox functions
try:
    from coach_toolbox_v2 import calculate_recommendation
except ImportError:
    # Fallback function if coach_toolbox_v2.py isn't present in current path
    def calculate_recommendation(stamina, passing):
        if stamina < 60:
            return "Increase stamina and rest prior to next match."
        elif passing < 80:
            return "Focus on passing drills."
        return "Optimal condition! Ready for full 90 minutes."

st.set_page_config(page_title="Soccer AI Coach", page_icon="⚽", layout="wide")

st.title("⚽ AI Soccer Coach Dashboard")
st.markdown("---")

# Load Data
df = pd.read_csv("Artin_FC.csv")

# Player Selection
selected_player = st.selectbox("Select Player:", df["name"].tolist())

# Fetch Selected Row
player_info = df[df["name"] == selected_player].iloc[0]

st.subheader(f"Player Statistics: {selected_player}")

# Multi-column Stat Metrics
col1, col2, col3 = st.columns(3)
col1.metric("Speed", player_info["speed"])
col2.metric("Stamina", player_info["stamina"])
col3.metric("Passing", player_info["passing"])

# Visual Stat Bar Chart
chart_data = pd.DataFrame({
    "Attribute": ["Speed", "Stamina", "Passing", "Shooting", "Defending"],
    "Rating": [
        player_info["speed"],
        player_info["stamina"],
        player_info["passing"],
        player_info["shooting"],
        player_info["defending"]
    ]
}).set_index("Attribute")

st.bar_chart(chart_data)

st.markdown("---")

# Connect AI Logic from coach_toolbox_v2.py
st.subheader("🤖 Coach Recommendation")
recommendation = calculate_recommendation(player_info["stamina"], player_info["passing"])
st.success(f'"{recommendation}"')

if st.button("Analyze Match Readiness"):
    st.balloons()
````
