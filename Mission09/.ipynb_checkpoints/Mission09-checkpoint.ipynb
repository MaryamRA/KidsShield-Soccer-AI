## 🔄 What Changes?

Before Mission 09, our code communicated through console logs and static notebook outputs. Now, Streamlit replaces those touchpoints with visual interface components.

| Before Mission 09 (Notebook) | After Mission 09 (Web App) |
| :--- | :--- |
| `print()` | `st.write()` / `st.markdown()` |
| `input()` | `st.text_input()` / `st.number_input()` |
| `display(df)` | `st.dataframe(df)` |
| `plt.show()` | `st.line_chart()` / `st.bar_chart()` / `st.pyplot()` |
| `if` statements | **Same `if` statements** |
| Custom functions | **Same custom functions** |
| `coach_toolbox_v2` | **Same `coach_toolbox_v2`** |
| `pandas` DataFrames | **Same `pandas` DataFrames** |

> 💡 **Notice something interesting?** Almost everything stays exactly the same. Streamlit only changes how information is displayed to the user.

## 📖 Python → Streamlit Translation Guide

Use this reference table whenever you want to convert Python terminal code into web app components:

| What You Know in Python | Streamlit Equivalent | Description |
| :--- | :--- | :--- |
| `print("Hello")` | `st.write("Hello")` / `st.text("Hello")` | Displays general text or variables |
| `print("# Title")` | `st.title("Title")` / `st.header("Header")` | Displays large headers |
| `input("Name: ")` | `st.text_input("Name: ")` | Text entry field |
| `int(input("Age: "))` | `st.number_input("Age: ")` | Numerical counter input |
| `input("Position: ")` | `st.selectbox("Position:", ["Forward", ...])` | Dropdown selection menu |
| `bool_val = True` | `st.checkbox("Enable Option")` | On/Off toggle switch |
| `choice = "A"` | `st.radio("Select Tactics:", ["A", "B"])` | Radio button selector |
| `display(df)` | `st.dataframe(df)` / `st.table(df)` | Interactive data tables |
| `plt.bar()` | `st.bar_chart(df)` | Renders interactive bar charts |
| `print("Error!")` | `st.error("Error!")` | Red alert message box |
| `print("Success!")` | `st.success("Success!")` | Green success alert box |
| `print("Warning!")` | `st.warning("Warning!")` | Yellow warning alert box |
| `print("Info!")` | `st.info("Info!")` | Blue informational alert box |

## 🧠 How Streamlit Thinks

In a standard Jupyter Notebook, code executes sequentially from top to bottom cell by cell:
```text
Notebook: Run Cell 1 ➔ Run Cell 2 ➔ Run Cell 3 (State updates manually)
```

**Streamlit uses a reactive model.** Every time a user clicks a button, moves a slider, or changes a dropdown, Streamlit reruns your Python script from line 1 to the end:

```text
[User Interacts with Widget]
          │
          ▼
┌─────────────────────────────────────────┐
│ Streamlit executes app.py from line 1   │
│ Reads current widget states             │
│ Re-evaluates all Python logic           │
│ Repaints the Web Dashboard in browser   │
└─────────────────────────────────────────┘
```

> 💡 **Key Mental Model:** You don't write code to handle DOM events. You just write simple, top-to-bottom Python logic. Streamlit handles redrawing the screen automatically!

## 🚀 Phase 1: Re-imagining Mission 1 — Variables & Displays

In Mission 1, we defined player profiles using variables and printed them. Let's make them visual!

### 💡 Think Like an AI Engineer
> *Why display static text when you can use structured UI elements like `st.metric()`? Metrics give coaches instant visual feedback on key performance indicators (KPIs).*

````python
import streamlit as st

# Mission 1 Code Concept: Player Variables
player_name = "Artin"
player_position = "Forward"
player_stamina = 90
player_speed = 92

# Streamlit Translation
st.title(f"⚽ Player Profile: {player_name}")
st.write(f"**Position:** {player_position}")

# Interactive KPIs
col1, col2 = st.columns(2)
col1.metric(label="Stamina Rating", value=f"{player_stamina}%", delta="+2%")
col2.metric(label="Speed Rating", value=f"{player_speed}/100", delta="+5%")
````

## 🚀 Phase 2: Re-imagining Mission 2 — Datasets & DataFrames

In Mission 2, we loaded the `Artin_FC` squad dataset. Let's make it interactive for coaches.

````python
import pandas as pd
import streamlit as st

# Mission 2 Dataset
squad_data = {
    "Name": ["Artin", "Messi", "Maradona", "Zidane", "Iniesta"],
    "Position": ["Forward", "Forward", "Midfielder", "Midfielder", "Midfielder"],
    "Stamina": [90, 85, 82, 84, 84],
    "Speed": [92, 89, 88, 80, 70],
    "Passing": [84, 85, 93, 83, 77]
}

df = pd.DataFrame(squad_data)

st.header("📊 Squad Roster")
st.dataframe(df, use_container_width=True)
````

## 🚀 Phase 3: Re-imagining Mission 3 — Condition Logic & Alert Boxes

In Mission 3, we wrote decision rules using `if/elif/else`. Let's translate console messages into colorful web alerts.

````python
import streamlit as st

st.header("🧠 Match Readiness Decision Engine")

# Interactive input widget
stamina_input = st.slider("Select Player Stamina:", min_value=30, max_value=100, value=55)

# Mission 3 Decision Rules converted to Streamlit alerts
if stamina_input < 60:
    st.error("🚨 CRITICAL WARNING: Player stamina too low! Bench player.")
elif stamina_input < 80:
    st.warning("⚠️ CAUTION: Player is sub-optimal. Limit playing time.")
else:
    st.success("✅ OPTIMAL FITNESS: Player is match-ready for full 90 minutes!")
````

---
## 🚀 Phase 4: Re-imagining Mission 4 — Integrating `coach_toolbox_v2`

In Mission 4, we created `coach_toolbox_v2.py`. In Streamlit, custom library imports work identically!

````python
import streamlit as st

# Stand-in function representing coach_toolbox_v2 module
def calculate_recommendation(stamina, passing):
    if stamina < 60:
        return "Rest player and focus on endurance recovery."
    elif passing < 80:
        return "Assign player to precision passing drills."
    return "Player in peak condition for match day!"

st.header("🧰 Coach Toolbox Integration")
p_stamina = st.number_input("Enter Player Stamina:", value=85)
p_passing = st.number_input("Enter Player Passing:", value=75)

rec = calculate_recommendation(p_stamina, p_passing)
st.info(f"💡 AI Coach Advice: {rec}")
````

In Mission 6, we filtered DataFrames using boolean masks. Let's attach a position dropdown directly to our filtering logic.

## 🚀 Phase 8: Re-imagining Mission 8 — Machine Learning Prediction Engine

In Mission 8, we trained ML models. Let's attach a live trained prediction engine to an interactive button!

````python
import streamlit as st

st.header("🤖 ML Match Performance Predictor")

st.write("Simulate AI prediction based on player attributes:")
st_val = st.slider("Stamina:", 50, 100, 80)
sp_val = st.slider("Speed:", 50, 100, 90)

if st.button("Run AI Prediction"):
    # Simulated model output from Mission 8
    score_pred = (st_val * 0.4) + (sp_val * 0.6)
    st.success(f"🎯 Projected Match Performance Score: {score_pred:.1f} / 100")
    st.balloons()
````

## 🏆 Capstone Project: Build the AI Soccer Coach Dashboard

Now, let's assemble everything from Missions 1–8 into a complete, professional multi-column dashboard!

### Directory Structure:
```text
soccer_ai_project/
│
├── coach_toolbox_v2.py   # Helper functions
├── Artin_FC.csv          # Squad dataset
└── app.py                # Final Streamlit app
```

````python
# Complete Streamlit Production Script (app.py)
import streamlit as st
import pandas as pd

# 1. Page Configuration
st.set_page_config(page_title="Artin FC AI Dashboard", page_icon="⚽", layout="wide")

# 2. Header Banner
st.title("⚽ Artin FC — AI Soccer Coach Dashboard")
st.markdown("*Real-time squad analytics, tactical decision engines, and ML performance projections.*")
st.markdown("---")

# 3. Dataset Initialization
squad_data = [
    {"Name": "Artin", "Position": "Forward", "Stamina": 90, "Speed": 92, "Passing": 84, "Shooting": 88},
    {"Name": "Messi", "Position": "Forward", "Stamina": 85, "Speed": 89, "Passing": 85, "Shooting": 96},
    {"Name": "Jahanbakhsh", "Position": "Forward", "Stamina": 80, "Speed": 85, "Passing": 68, "Shooting": 80},
    {"Name": "Maradona", "Position": "Midfielder", "Stamina": 82, "Speed": 88, "Passing": 93, "Shooting": 92},
    {"Name": "Zidane", "Position": "Midfielder", "Stamina": 84, "Speed": 80, "Passing": 83, "Shooting": 84},
    {"Name": "Iniesta", "Position": "Midfielder", "Stamina": 84, "Speed": 70, "Passing": 77, "Shooting": 74},
    {"Name": "Maldini", "Position": "Defender", "Stamina": 85, "Speed": 82, "Passing": 83, "Shooting": 60},
    {"Name": "Beiranvand", "Position": "GK", "Stamina": 75, "Speed": 65, "Passing": 63, "Shooting": 10}
]
df = pd.DataFrame(squad_data)

# 4. Top Key Metrics Row
kpi1, kpi2, kpi3, kpi4 = st.columns(4)
kpi1.metric("Total Squad Size", len(df))
kpi2.metric("Avg Team Stamina", f"{df['Stamina'].mean():.1f}")
kpi3.metric("Avg Team Speed", f"{df['Speed'].mean():.1f}")
kpi4.metric("Top Scorer Rating", f"{df['Shooting'].max()}")

st.markdown("---")

# 5. Main Dashboard Layout (2 Columns)
col_left, col_right = st.columns([1, 1])

with col_left:
    st.subheader("📋 Squad Roster & Position Filter")
    pos_filter = st.selectbox("Filter Position:", ["All"] + list(df["Position"].unique()))
    if pos_filter != "All":
        filtered_df = df[df["Position"] == pos_filter]
    else:
        filtered_df = df
    st.dataframe(filtered_df, use_container_width=True)

with col_right:
    st.subheader("👤 Individual Player Profile")
    selected_player = st.selectbox("Select Player:", df["Name"].tolist())
    player_data = df[df["Name"] == selected_player].iloc[0]
    p_col1, p_col2, p_col3 = st.columns(3)
    p_col1.metric("Stamina", player_data["Stamina"])
    p_col2.metric("Speed", player_data["Speed"])
    p_col3.metric("Passing", player_data["Passing"])
    p_chart = pd.DataFrame({
        "Attribute": ["Stamina", "Speed", "Passing", "Shooting"],
        "Value": [player_data["Stamina"], player_data["Speed"], player_data["Passing"], player_data["Shooting"]]
    }).set_index("Attribute")
    st.bar_chart(p_chart)

st.markdown("---")

# 6. AI Match Decision Engine
st.subheader("🤖 AI Match Decision Engine")
col_ai1, col_ai2 = st.columns(2)

with col_ai1:
    st.markdown("### Match Fitness Rules")
    if player_data["Stamina"] < 80:
        st.warning(f"⚠️ {selected_player} needs rest management.")
    else:
        st.success(f"✅ {selected_player} is fully match-ready.")

with col_ai2:
    st.markdown("### ML Performance Projection")
    if st.button(f"Predict Match Score for {selected_player}"):
        pred_score = (player_data["Speed"] * 0.5) + (player_data["Shooting"] * 0.5)
        st.info(f"Predicted Impact Rating: {pred_score:.1f} / 100")
        st.balloons()
````
