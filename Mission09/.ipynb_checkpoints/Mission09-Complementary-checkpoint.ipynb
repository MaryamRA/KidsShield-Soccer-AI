## 1. Threat Matrix & Risk Calculation

Risk is commonly modeled as:
$$\text{Risk} = \text{Likelihood} \times \text{Impact}$$

Below is a simple helper function to quantify threat levels based on CVSS-style score inputs.

```python
def calculate_risk(likelihood: int, impact: int) -> dict:
    """
    Calculates a simple risk score given likelihood and impact on a 1-5 scale.
    """
    if not (1 <= likelihood <= 5 and 1 <= impact <= 5):
        raise ValueError("Likelihood and Impact must be integers between 1 and 5.")
        
    score = likelihood * impact
    
    if score >= 20:
        severity = "CRITICAL"
    elif score >= 12:
        severity = "HIGH"
    elif score >= 6:
        severity = "MEDIUM"
    else:
        severity = "LOW"
        
    return {"score": score, "severity": severity}

# Test calculation
sample_risk = calculate_risk(likelihood=4, impact=4)
print(f"[+] Threat Assessment Output: {sample_risk}")
```

## 2. Practical Exercise

Use the cell below to analyze potential vectors for your scenario and log the results.

```python
threats = [
    {"vector": "Credential Stuffing", "likelihood": 4, "impact": 3},
    {"vector": "Phishing / Social Engineering", "likelihood": 5, "impact": 4},
    {"vector": "Unpatched Dependency Vulnerability", "likelihood": 3, "impact": 5}
]

print("--- THREAT LOG ---")
for t in threats:
    assessment = calculate_risk(t["likelihood"], t["impact"])
    print(f"Vector: {t['vector']:<35} | Score: {assessment['score']:<2} | Severity: {assessment['severity']}")
```
