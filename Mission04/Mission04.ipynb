## 🟩 Part 1 — Why Functions Exist
Without functions, if you want to find a player's performance contribution across 20 different players, you would have to manually type or copy-paste code like this:
```python
print(player["goals"] + player["assists"])
```

With functions, we use the `def` keyword (short for *define*) to create a machine template once, and reuse it anywhere!

````python
# Defining our very first soccer tool template
def total_contribution(goals, assists):
    return goals + assists

# Testing the tool
messi_contribution = total_contribution(12, 6)
#print("Messi Total Contribution:", messi_contribution)
````

## 🟩 Part 2 — Parameters and Arguments

````python
# A real dictionary representing our player data row
player_1 = {
    "name": "Isaac",
    "position": "Forward",
    "goals": 12,
    "assists": 6
}

def total_contribution_dict(player):
    # The 'player' parameter acts as a doorway for our dictionary
    return player["goals"] + player["assists"]

# Passing the whole player into the tool
result = total_contribution_dict(player_1)
print(f"{player_1['name']} total output value:", result)
````

## 🟩 Part 3 — Return Values vs. Printing
There is a huge distinction between `print()` and `return`:
- `print()` simply splashes text out onto the screen for human eyes. The computer immediately forgets it.
- `return` passes the calculated value **back into the computer's memory bank**, allowing you to save it into variables to use in calculations later!

````python
def get_rating_value():
    return 85

# Because it uses 'return', we can save it and do math on it later!
saved_rating = get_rating_value()
boosted_rating = saved_rating + 5
print("Boosted Rating Memory State:", boosted_rating)
````

### 📚 What is a Code Library?
Up until now, you have been building individual tools (functions) one by one. But as your software grows, keeping all your tools loose in your workspace makes things messy.

In programming, a Library is a collection of reusable code (like functions, variables, or classes) that are organized together into a single package.

The Toolbox Analogy: Think of individual functions as tools—a hammer, a screwdriver, or a wrench. A Library is the actual heavy-duty toolbox where you organize and store all those tools so you can carry them to any soccer stadium or project easily without losing them.

💡 A Simple Example of a Library
In Python, you can group functions together into a .py file (called a module) and then bring them into your main notebook using the import command.

Imagine we have a mini-library file called soccer_math.py:

````python
# Content inside soccer_math.py
def calculate_points(wins, draws):
    return (wins * 3) + (draws * 1)
````

To use this library in our main project, we simply import it like this:

````python
import soccer_math

# Using our library tool
team_points = soccer_math.calculate_points(wins=5, draws=2)
print(f"Total Team Points: {team_points}")
````
