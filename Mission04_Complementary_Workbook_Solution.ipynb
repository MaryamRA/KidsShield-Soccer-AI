You can develop your calculate_efficiency(player_profile) tool like the following:

```python
def calculate_efficiency(player_profile):
    total_comp = 0
    total_att = 0

    # Loop through tracking logs
    for cp in player_profile["passes_completed"]:
        total_comp = total_comp + cp

    for tp in player_profile["total_passes"]:
        total_att = total_att + tp

    # Standard precision math division
    accuracy_percentage = (total_comp / total_att) * 100


    return accuracy_percentage

# 🧪 TEST YOUR EFFICIENCY ENGINE
artin_efficiency = calculate_efficiency(target_player)
print("Efficiency Rating:", artin_efficiency, "%")
```
