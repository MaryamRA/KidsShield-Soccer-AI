## 🧠 Part 3 – Your First Decision Engine

```
def coach_decision(player):

    if coach_toolbox.needs_rest(player):
        return "Recovery Day"

    elif coach_toolbox.needs_endurance_training(player):
        return "Endurance Training"

    elif coach_toolbox.needs_shooting_practice(player):
        return "Shooting Practice"

    else:
        return "Ready for Match"
```

### Exercise
Test your decision engine on one player.

```
print(players[0]["name"])
print(coach_decision(players[0]))
```

## 📋 Part 4 – Build a Player Report

```
def create_player_report(player):

    return {
        "name": player["name"],
        "position": player["position"],
        "needs_rest": coach_toolbox.needs_rest(player),
        "excellent_stamina": coach_toolbox.excellent_stamina(player),
        "needs_shooting_practice": coach_toolbox.needs_shooting_practice(player),
        "recommendation": coach_decision(player)
    }
```

```
report = create_player_report(players[0])
report
```

## 👥 Part 5 – Analyze the Whole Team

```
reports=[]

for player in players:
    reports.append(create_player_report(player))

reports[:3]
```

## 🐼 Part 6 – Meet pandas

```
df = pd.DataFrame(reports)

df.head()
```

Explore your data.

```
df.info()
df.describe(include="all")
```

## 📊 Part 7 – Simple Visualizations

```
df["recommendation"].value_counts().plot(kind="bar")

plt.title("Coaching Recommendations")
plt.xlabel("Recommendation")
plt.ylabel("Players")
plt.show()
```

```
df["position"].value_counts().plot(kind="pie", autopct="%1.0f%%")

plt.ylabel("")
plt.title("Player Positions")
plt.show()
```

## 📈 Part 8 – Team Summary

```
print("Players:", len(df))
print("Need Rest:", df["needs_rest"].sum())
print("Excellent Stamina:", df["excellent_stamina"].sum())
print("Need Shooting Practice:", df["needs_shooting_practice"].sum())
```
