## 🕸️ Part 3: Advanced Player Radar Charts (Skill Signatures)

A **Radar (Spider) Chart** is the ultimate way to visualize a player's raw strength footprint, just like in FIFA or EA Sports FC!

Let's write a function to plot and compare **Artin's** skill blueprint against the legendary **Messi**!

```python
import numpy as np

def plot_player_radar(player1_name, player2_name):
    # 1. Define the skill categories we want to plot
    categories = ['passing', 'defending', 'stamina', 'speed', 'shooting']
    num_vars = len(categories)
    
    # 2. Get the values for both players
    p1_data = df.loc[df['name'] == player1_name, categories].values[0].tolist()
    p2_data = df.loc[df['name'] == player2_name, categories].values[0].tolist()
    
    # 3. Close the radar loop by repeating the first value at the end
    p1_data += p1_data[:1]
    p2_data += p2_data[:1]
    
    # 4. Calculate angle for each axis
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
    angles += angles[:1]
    
    # 5. Initialize the polar plot
    fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
    
    # Draw Player 1
    ax.plot(angles, p1_data, color='blue', linewidth=2, label=player1_name)
    ax.fill(angles, p1_data, color='blue', alpha=0.25)
    
    # Draw Player 2
    ax.plot(angles, p2_data, color='orange', linewidth=2, label=player2_name)
    ax.fill(angles, p2_data, color='orange', alpha=0.25)
    
    # Customize Labels
    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)
    plt.xticks(angles[:-1], categories, fontsize=12, fontweight="bold")
    
    plt.title(f"⚔️ Skill Signature Battle: {player1_name} vs. {player2_name}", fontsize=16, fontweight="bold", pad=20)
    plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1.1))
    plt.show()

# Run the combat profile!
plot_player_radar("Artin", "Messi")
```

## 🌡️ Part 4: Skills Correlation Heatmap

How do skills relate to each other? Does a high speed rating usually mean a player has high stamina? 

We can calculate a **correlation matrix** and visualize it as a colorful thermal **Heatmap**.

```python
plt.figure(figsize=(8, 6))

# Extract numerical skill columns
skills_df = df[['passing', 'defending', 'stamina', 'speed', 'shooting']]
corr_matrix = skills_df.corr()

# Plot the Heatmap with values annotated inside the squares
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=1.5, vmin=-1, vmax=1)

plt.title("🌡️ Player Attribute Correlation Heatmap", fontsize=15, fontweight="bold")
plt.show()
```

---

## 🧬 Part 5: Skill Distributions (Joint Scatter Plots)

Let's look at **Passing vs. Shooting** using a `jointplot`! This gives us a scatter plot of player coordinates *and* showing individual distribution histograms on the outer edges simultaneously.

```python
# Create a joint distribution map to see patterns
g = sns.jointplot(x="passing", y="shooting", data=df, kind="reg", color="purple", height=7)
g.fig.suptitle("🎯 Joint Attribute Analysis: Passing vs. Shooting", y=1.03, fontsize=14, fontweight="bold")
plt.show()
```

## 🏆 Boss Challenge: Tactical Dashboard

Create a dashboard featuring **two side-by-side plots** on one canvas:
- **Left Plot**: A sorted barplot comparing `shooting` performance.
- **Right Plot**: A `passing` versus `defending` scatter plot, colored by their `position`.

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 7))

# --- Left Plot: Shooting Bar Chart ---
df_sorted_shooting = df.sort_values("shooting", ascending=False)
sns.barplot(ax=axes[0], x="name", y="shooting", data=df_sorted_shooting, palette="rocket")
axes[0].set_title("🎯 Shooting Performance Index", fontsize=14, fontweight="bold")
axes[0].set_xticklabels(df_sorted_shooting["name"], rotation=45)

# --- Right Plot: Passing vs Defending Tactical Scatter ---
sns.scatterplot(ax=axes[1], x="passing", y="defending", hue="position", style="position", s=200, data=df)
axes[1].set_title("🛡️ Playmaking (Passing) vs. Defensive Capabilities", fontsize=14, fontweight="bold")

plt.tight_layout()
plt.show()
```
