### 🧩 Challenge 2.1: The Custom Overall Rating Calculator
**Goal:** Let's define an updated overall rating function that calculates a player's total performance. 
For outfield players (everyone except the goalkeeper 'GK'), their overall score is the mathematical average of all 5 skill attributes.
For Goalkeepers ('GK'), defending and shooting aren't as important, so their overall rating is calculated differently: `(passing * 0.40) + (stamina * 0.40) + (speed * 0.20)`.

````python
def calculate_overall_rating_v2(player):
    if player["position"] == "GK":
        rating = (player["passing"] * 0.40) + (player["stamina"] * 0.40) + (player["speed"] * 0.20)
    else:
        total_skills = player["passing"] + player["defending"] + player["stamina"] + player["speed"] + player["shooting"]
        rating = total_skills / 5
    return round(rating, 1)

# Let's test our formula
print("Artin Overall Skill Rating:", calculate_overall_rating_v2(Artin_FC_v2[0]), "%")
print("Beiranvand GK Skill Rating:", calculate_overall_rating_v2(Artin_FC_v2[-1]), "%")
````

## 📈 Milestone 3: Enter Pandas & Data Science Visualizations

Instead of looking at dictionaries filled with text and brackets, professional data scientists use a library called **Pandas** to turn data lists into spreadsheet tables called **DataFrames**.

````python
import pandas as pd

# Let's convert our list of player dictionaries into a Pandas DataFrame
df = pd.DataFrame(Artin_FC_v2)

# Let's display the table!
df
````

### 🎨 Generating Cool Charts and Visualizations
We can use plotting libraries to turn this structured DataFrame table into instant graphs to compare our player statistics visually!

````python
import matplotlib.pyplot as plt

# Let's generate a vertical bar chart comparing the speed and shooting of all players!
plt.figure(figsize=(10, 5))
plt.bar(df['name'], df['speed'], color='skyblue', label='Speed')
plt.bar(df['name'], df['shooting'], color='salmon', label='Shooting', alpha=0.7)

plt.title('⚡ Speed vs. Shooting Performance Comparison', fontsize=14, fontweight='bold')
plt.xlabel('Player Name', fontsize=12)
plt.ylabel('Skill Rating (Out of 100)', fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

## 🧠 Milestone 4: Building the Intelligent Decision Agent

Now we are going to combine all our toolbox measurements to make tactical decisions. An **AI Decision Function** analyzes data points and outputs real coaching recommendations.

### 🧩 Challenge 4.1: The Tactical AI Agent recommendation Engine
Let's write a function that reviews a player dictionary and returns a tailored coaching card:
- If a player's defending attribute is greater than 80, recommend: `"Solid defensive anchor. Hold the backline."`
- If their speed is greater than 90 and passing is greater than 80, recommend: `"Winger breakout threat. Focus on flank overlaps."`
- If their shooting is greater than 90, recommend: `"Target striker. Look for direct shots inside the box."`
- Otherwise, return: `"Tactical supporting playmaker. Distribute cleanly."`

````python
def generate_ai_tactical_recommendation(player):
    if player["defending"] >= 80:
        return "🛡️ Solid defensive anchor. Hold the backline."
    elif player["speed"] >= 90 and player["passing"] >= 80:
        return "⚡ Winger breakout threat. Focus on flank overlaps."
    elif player["shooting"] >= 90:
        return "⚽ Target striker. Look for direct shots inside the box."
    else:
        return "🎯 Tactical supporting playmaker. Distribute cleanly."

# Let's run a test
for player in Artin_FC_v2:
    print(f"👤 Player: {player['name']} ({player['position']})")
    print(f"   Recommendation: {generate_ai_tactical_recommendation(player)}")
    print("-----------------------------------------------------------")
````
