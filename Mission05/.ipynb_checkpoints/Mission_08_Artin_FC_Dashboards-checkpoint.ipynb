## 👁️ Why Coaches Use Dashboards

In elite sports, data visualization is used for:
1. **Cognitive Efficiency:** The human brain processes images up to 60,000 times faster than text. A color-coded chart reveals a player's physical fatigue or shooting accuracy in a fraction of a second.
2. **Patterns Over Rows:** Comparing 11 players across 5 stats creates 55 data points. A scatter plot or bar chart condenses those 55 points into a single intuitive picture.
3. **Evidence-Based Strategy:** When explaining to a player why they are starting on the bench or changing positions, showing them a visual trend of their performance builds trust and clarity.

## 🎨 2. Introduction to Matplotlib

**Matplotlib** is Python's foundational plotting library. It operates like an artist's canvas: 
1. You declare a figure frame (`plt.figure()`).
2. You draw elements onto it (bars, lines, scatter points).
3. You add labels and guidelines (`plt.xlabel()`, `plt.title()`, `plt.grid()`).
4. You render the image to the screen (`plt.show()`).

**Seaborn** sits on top of Matplotlib, automatically applying crisp presets, modern color-grading curves (palettes like `viridis`, `rocket`, `crest`), and automatically setting titles and category boundaries.

## 🛠️ 3. Building Your First Chart Step-by-Step

Let's construct a chart piece by piece so you can see how each line of code acts as a brush stroke.

```python
# Step 1: Initialize the Canvas (Width=10 inches, Height=5 inches)
plt.figure(figsize=(10, 5))

# Step 2: Draw the graphic element using Seaborn
sns.barplot(x="name", y="speed", data=df, palette="Blues_r")

# Step 3: Decorate the plot with Labels and titles
plt.title("⚡ Squad Speed Index (Descending)", fontsize=14, fontweight="bold")
plt.xlabel("Player Roster", fontsize=11)
plt.ylabel("Speed Rating", fontsize=11)

# Step 4: Rotate categories so names do not overlap
plt.xticks(rotation=45)

# Step 5: Render and clear layout memory buffers
plt.tight_layout()
plt.show()
```

### 💡 Dashboard 7 – Recommendation Breakdown
**Tactical Goal:** Sort players into customized recommendations based on their key profiles. 
Let's write a simple selector function and visualize our lineup categories!

```python
def attribute_classifier(row):
    if row['shooting'] >= 85:
        return "Elite Finisher"
    elif row['passing'] >= 85 and row['defending'] < 60:
        return "Creative Playmaker"
    elif row['defending'] >= 80:
        return "Defensive Anchor"
    else:
        return "Utility Reserve"

df['tactical_profile'] = df.apply(attribute_classifier, axis=1)

plt.figure(figsize=(8, 5))
sns.countplot(y="tactical_profile", data=df, palette="pastel", order=df['tactical_profile'].value_counts().index)
plt.title("💡 Roster Tactical Profile Distribution", fontsize=14, fontweight="bold", pad=15)
plt.xlabel("Count", fontsize=11)
plt.ylabel("Assigned Role Profile", fontsize=11)
plt.tight_layout()
plt.show()
```

### ⏳ Dashboard 8 – Age vs. Goals (Scatter Plot)
**Tactical Goal:** Do we rely too heavily on older players for goals? Let's check the age-vs-goals distribution.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="age", y="total_goals", hue="position", style="position", s=180, data=df)
plt.title("⏳ Player Age vs. Goals Over Season", fontsize=14, fontweight="bold", pad=15)
plt.xlabel("Player Age (Years)", fontsize=11)
plt.ylabel("Goals Scored", fontsize=11)
plt.tight_layout()
plt.show()
```

### 🔋 Dashboard 9 – Stamina vs. Goals (Scatter Plot)
**Tactical Goal:** Do the hardest-working physical players also convert those movements into goals?

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="stamina", y="total_goals", hue="position", size="speed", sizes=(40, 240), data=df)
plt.title("🔋 Stamina vs. Goals Scored (Size = Speed Rating)", fontsize=14, fontweight="bold", pad=15)
plt.xlabel("Stamina Rating", fontsize=11)
plt.ylabel("Goals Scored", fontsize=11)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### 🏟️ Dashboard 10 – Team Performance Summary
**Tactical Goal:** Construct a single multi-panel chart comparing three critical technical ratings: `passing`, `defending`, and `shooting` side-by-side using violin distributions.

```python
melted_skills = pd.melt(df, id_vars=['name'], value_vars=['passing', 'defending', 'shooting'], 
                        var_name='Attribute', value_name='Rating')

plt.figure(figsize=(10, 6))
sns.violinplot(x="Attribute", y="Rating", data=melted_skills, inner="point", palette="coolwarm")
plt.title("🏟️ Team Core Skills Signature Distributions", fontsize=14, fontweight="bold", pad=15)
plt.xlabel("Tactical Dimension", fontsize=11)
plt.ylabel("Attribute Rating Scale (0-100)", fontsize=11)
plt.tight_layout()
plt.show()
```

## 🛠️ 5. Building the Complete Team Dashboard Function

Let's pack our data visualizations into a single reusable dashboard function that prints a beautifully formatted scouting summary.

```python
def show_team_dashboard(dataframe):
    """
    Generates a unified tactical summary console containing key stats
    and structural breakdowns for Artin FC.
    """
    print("=" * 60)
    print("⚽ ARTIN FC SYSTEM SCOUTING CONSOLE ⚽")
    print("=" * 60)
    
    # 1. Compute summary indicators
    total_squad_goals = dataframe['total_goals'].sum()
    avg_passing_accuracy = dataframe['passing_accuracy'].mean()
    top_contributor = dataframe.loc[dataframe['goal_contributions'].idxmax(), 'name']
    
    print(f"🔹 Total Seasonal Team Goals: {total_squad_goals}")
    print(f"🔹 Average Squad Passing Efficiency: {avg_passing_accuracy:.2f}%")
    print(f"🔹 Star Offensive Playmaker: {top_contributor}")
    print("-" * 60)
    
    # 2. Build multi-panel grid plot
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    
    # Panel A: Sorted Direct Contributions
    sorted_df = dataframe.sort_values("goal_contributions", ascending=False)
    sns.barplot(ax=axes[0], x="name", y="goal_contributions", data=sorted_df, palette="crest")
    axes[0].set_title("Direct Goal Contributions (Goals + Assists)", fontsize=12, fontweight="bold")
    axes[0].tick_params(axis='x', rotation=45)
    
    # Panel B: Defensive vs Passing Coordinates
    sns.scatterplot(ax=axes[1], x="passing", y="defending", hue="position", s=180, data=dataframe)
    axes[1].set_title("Positioning Map: Defense vs. Passing Abilities", fontsize=12, fontweight="bold")
    
    plt.tight_layout()
    plt.show()

# Deploy our Dashboard System!
show_team_dashboard(df)
```
