## 🧱 Part 1: Setting Up the Art Studio (Matplotlib & Seaborn)

To make plots in Python, we use two incredible libraries:
1. **`matplotlib.pyplot`**: The core drafting tool. We use the abbreviation **`plt`** to make our code faster to write.
2. **`seaborn`**: The style master. It automatically formats standard charts to look modern and adds advanced color choices. We abbreviate it as **`sns`**.

Let's import them and apply a stylish dark or light theme for our coach's dashboard!

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the default stylish theme for Seaborn
sns.set_theme(style="whitegrid")
print("🎨 Visualization tools imported and themed!")
```

---

## 📊 Part 2: Building Bar Charts (Comparing Player Stamina)

Our first visualization is the **Bar Chart**. Bar charts are perfect for comparing different players' values head-to-head.

Let's write a program that plots everyone's name on the horizontal bottom axis ($X$-axis) and their stamina value on the vertical side axis ($Y$-axis).

### 💡 Anatomy of a Plot in Python:
- `plt.figure(figsize=(width, height))`: Tells Python how wide and tall to build our blank canvas.
- `sns.barplot(x=..., y=..., data=df)`: Maps columns directly to the $X$ and $Y$ coordinates.
- `plt.title()`: Adds a header title.
- `plt.xlabel()` and `plt.ylabel()`: Label our chart borders.
- `plt.xticks(rotation=45)`: Tilts our player names sideways so they don't overlap!
- `plt.show()`: Tells Python to render the final image on our screen.

```python
# 1. Setup Canvas Size (12 units wide, 6 units tall)
plt.figure(figsize=(12, 6))

# 2. Draw the Barplot with a 'viridis' gradient color palette
sns.barplot(x="name", y="stamina", data=df, palette="viridis")

# 3. Add Labels & Style
plt.title("🫁 Artin FC: Player Stamina Comparison", fontsize=16, fontweight="bold")
plt.xlabel("Player Name", fontsize=12)
plt.ylabel("Stamina Rating", fontsize=12)

# 4. Rotate player names 45 degrees so we can read them easily!
plt.xticks(rotation=45)

# 5. Output the masterpiece!
plt.show()
```

## 🎯 Part 3: Scatter Plots (Is Speed Related to Shooting?)

A **Scatter Plot** is a data scientist's favorite tool for investigating relationships between two stats. 

For example, let's investigate: **Are our fastest players also our best shooters?**

We can place `speed` on the $X$-axis and `shooting` on the $Y$-axis. Each player will become a single dot on the grid. To make it even cooler, we can use the `hue` option to color the dots based on their active soccer `position`!

```python
plt.figure(figsize=(10, 6))

# Draw scatter plot, coloring dots by player position
sns.scatterplot(x="speed", y="shooting", hue="position", style="position", s=150, data=df)

plt.title("⚡ Speed vs. Shooting Skill Relationship", fontsize=16, fontweight="bold")
plt.xlabel("Speed Rating", fontsize=12)
plt.ylabel("Shooting Rating", fontsize=12)

plt.show()
```

## 🏔️ Part 4: Distribution Plots (Understanding Team Age)

A **Distribution Plot** (often called a Histogram or KDE plot) shows us how spread out our team's demographics are. 

Let's check our roster's age brackets. Do we have a team of mostly 11-to-13-year-olds with a few adult players? We can view this spread using `sns.histplot` with an optional trend curve called a `kde`.

```python
plt.figure(figsize=(8, 5))

# Build the age distribution histogram with custom spacing bins
sns.histplot(df["age"], kde=True, color="teal", bins=10)

plt.title("🎂 Roster Age Distribution Spread", fontsize=15, fontweight="bold")
plt.xlabel("Player Age", fontsize=11)
plt.ylabel("Number of Players", fontsize=11)

plt.show()
```

---

## 📈 Part 5: Line Plots (Match-by-Match Goals Trend)

So far, we have only visualized totals. What if the coach wants to look at match history? 

For example, let's track the goalscoring performance of **Artin** and **Messi** across their last 4 matches. 

Because `goals` is currently saved as a list (e.g., `[2, 1, 0, 3]`), we can locate their lists using `.loc` and draw a custom match-by-match timeline chart!

```python
# 1. Select Artin's goals list using our .loc technique!
# (We find the row where 'name' == 'Artin', and select the 'goals' column)
artin_goals = df.loc[df['name'] == 'Artin', 'goals'].values[0]

# 2. Select Messi's goals list as well
messi_goals = df.loc[df['name'] == 'Messi', 'goals'].values[0]

# 3. Set our X-axis values for our match timeline labels
matches = ["Match 1", "Match 2", "Match 3", "Match 4"]

# 4. Plot the data canvas
plt.figure(figsize=(10, 5))

# Draw line charts with data dots ('marker') on each point
plt.plot(matches, artin_goals, marker='o', linewidth=3, label="Artin", color="royalblue")
plt.plot(matches, messi_goals, marker='s', linewidth=3, label="Messi", color="crimson")

plt.title("🎯 Match-by-Match Goal Scoring Performance", fontsize=15, fontweight="bold")
plt.xlabel("Match History Timeline", fontsize=12)
plt.ylabel("Goals Scored", fontsize=12)
plt.ylim(-0.5, 4.5)  # Set visual vertical limits on our graph
plt.legend()         # Display our colored player key

plt.show()
```
