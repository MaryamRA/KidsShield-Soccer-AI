## 🧱 Part 1: Meet Your Team Dashboard

Instead of managing a giant list of curly-braces and lists, let's load this dataset into the data science standard spreadsheet grid: a **Pandas DataFrame**.

```python
import pandas as pd

# Convert our player dictionaries list to a Pandas DataFrame
df = pd.DataFrame(Artin_FC_v2)

# Show the first few rows of our table!
df.head()
```

### 💡 Under the Hood: What is a DataFrame?
* **Rows:** Each row (horizontal) represents a single **individual player**.
* **Columns:** Each column (vertical) represents a specific **attribute or metric** of information about that player.

--- 

## 🔍 Part 2: Become a Data Detective

Let's write direct query calculations to answer immediate coaching questions about our player group:

```python
# 1. How many players do we have on the roster?
print("👥 Roster Size:", len(df))

# 2. What is the average stamina rating across our whole team?
print("🫁 Average Stamina:", df["stamina"].mean())

# 3. Who is our youngest and oldest player?
print("👶 Youngest Age:", df["age"].min())
print("🧓 Oldest Age:", df["age"].max())

# 4. What is the most goals scored in a single match by a player?
# (Since goals is a list of 4 match scores, let's look at who scored the highest individual total)
df['total_goals'] = df['goals'].apply(sum)
max_goals_idx = df['total_goals'].idxmax()
print(f"⚽ Top Goal Scorer: {df.loc[max_goals_idx, 'name']} with {df.loc[max_goals_idx, 'total_goals']} total goals!")

# 5. Who is our top assistant?
df['total_assists'] = df['assists'].apply(sum)
max_assists_idx = df['total_assists'].idxmax()
print(f"🎯 Top Playmaker: {df.loc[max_assists_idx, 'name']} with {df.loc[max_assists_idx, 'total_assists']} total assists!")
```

--- 

## 📈 Part 3: Build Your First Dashboard

Running `df.describe()` automatically computes a comprehensive analytical breakdown of every numerical column in your dataset!

```python
# Generate automated dataset overview metrics
df.describe()
```

### 📈 Decoding the Statistical Outputs
* **`count`**: The total count of valid items present in that column.
* **`mean`**: The mathematical average value of the column metrics.
* **`std`**: Standard Deviation (tells you how spread out the values are from the average. High deviation means the team has a wide range of skill ratings; low means they have uniform skill ranges).
* **`min`**: The minimum value recorded in that column.
* **`max`**: The maximum value recorded in that column.

--- 

## 🛡️ Part 4: Filter Like a Scout

Instead of looking at the whole team, let's isolate specific sub-groups of players using logical filters:

```python
# Filter: Only show players whose position is 'Forward'
forwards = df[df["position"] == "Forward"]
print("⚡ Forwards on Roster:")
print(forwards[["name", "speed", "shooting"]])
```

```python
# Filter: Only show defenders with speed higher than 80
fast_defenders = df[(df["position"] == "Defender") & (df["speed"] > 80)]
print("🛡️ Fast Defensive Lineups:")
print(fast_defenders[["name", "defending", "speed"]])
```

```python
# Filter: Only show players under age 13
junior_squad = df[df["age"] < 13]
print("👶 Junior Team Selection:")
print(junior_squad[["name", "age", "position"]])
```

--- 

## 🧬 Part 5: Add New Columns

We can expand our active DataFrame dynamically! Let's compute two major columns:
1. **Passing Accuracy:** The ratio of completed passes over total attempts.
2. **Needs Rest:** A boolean column that checks if their average stamina is falling low.

```python
# 1. Compute total completed passes over total passes
df["total_completed"] = df["passes_completed"].apply(sum)
df["total_attempts"] = df["total_passes"].apply(sum)
df["passing_accuracy"] = round((df["total_completed"] / df["total_attempts"]) * 100, 1)

# 2. Calculate if they need rest (if their active stamina rating is less than 82)
df["needs_rest"] = df["stamina"] < 82

# Let's view our updated DataFrame columns!
df[["name", "position", "passing_accuracy", "stamina", "needs_rest"]]
```

## 🎨 Part 6: Visualization Time

Let's create **10 interactive tactical visualizations**! Every single chart will solve a direct coaching mystery.

```python
import matplotlib.pyplot as plt
import numpy as np
```

### 📈 Dashboard 1: Goals by Player (Bar Chart)
**Question:** Who are our primary goal threats on the pitch?

```python
plt.figure(figsize=(10, 4))
plt.bar(df["name"], df["total_goals"], color="forestgreen", edgecolor="black")
plt.title("⚽ Total Goals Scored by Player", fontsize=13, fontweight="bold")
plt.ylabel("Goals Scored")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

### 📈 Dashboard 2: Assists by Player (Horizontal Bar Chart)
**Question:** Who is our primary playmaker setting up goals?

```python
plt.figure(figsize=(10, 4))
plt.barh(df["name"], df["total_assists"], color="gold", edgecolor="black")
plt.title("🎯 Total Assists Contributed by Player", fontsize=13, fontweight="bold")
plt.xlabel("Assists Made")
plt.tight_layout()
plt.show()
```

### 📈 Dashboard 3: Position Distribution (Pie Chart)
**Question:** Do we have a balanced squad lineup or are we missing defenders?

```python
plt.figure(figsize=(6, 6))
df["position"].value_counts().plot(kind="pie", autopct="%1.1f%%", colors=["skyblue", "orange", "lightgreen", "violet"], startangle=90)
plt.title("📋 Roster Tactical Position Balance", fontsize=13, fontweight="bold")
plt.ylabel("")
plt.show()
```

### 📈 Dashboard 4: Stamina Distribution (Histogram)
**Question:** What is the athletic fitness level of our club?

```python
plt.figure(figsize=(10, 4))
plt.hist(df["stamina"], bins=5, color="salmon", edgecolor="black", alpha=0.85)
plt.title("🫁 Squad Stamina Range Distribution", fontsize=13, fontweight="bold")
plt.xlabel("Stamina Attribute Value")
plt.ylabel("Number of Players")
plt.grid(axis="x", linestyle="--", alpha=0.5)
plt.show()
```

### 📈 Dashboard 5: Age vs Goals (Scatter Plot)
**Question:** Do older players on the squad tend to score more goals?

```python
plt.figure(figsize=(10, 4))
plt.scatter(df["age"], df["total_goals"], s=df["stamina"]*2, color="purple", alpha=0.7, edgecolor="black")
plt.title("🏃 Age vs. Goals Scored (Size = Stamina)", fontsize=13, fontweight="bold")
plt.xlabel("Player Age (Years)")
plt.ylabel("Total Goals Scored")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

### 📈 Dashboard 6: Passing Accuracy Ranking (Horizontal Bar Chart)
**Question:** Who is the best passer on the team?

```python
sorted_df = df.sort_values(by="passing_accuracy", ascending=True)

plt.figure(figsize=(10, 5))
plt.barh(sorted_df["name"], sorted_df["passing_accuracy"], color="teal", edgecolor="black")
plt.title("🎯 Passing Accuracy Performance Ranking", fontsize=13, fontweight="bold")
plt.xlabel("Passing Accuracy (%)")
plt.grid(axis="x", linestyle="--", alpha=0.5)
plt.show()
```

### 📈 Dashboard 7: Tactical Recommendations (Bar Chart)
**Question:** How does the team sort out across tactical recommendations?

```python
def resolve_status(player):
    if player["needs_rest"]:
        return "Recovery"
    elif player["stamina"] < 85:
        return "Endurance Practice"
    elif player["shooting"] < 80:
        return "Shooting Practice"
    else:
        return "Match Ready"

df["status"] = df.apply(resolve_status, axis=1)

plt.figure(figsize=(8, 4))
df["status"].value_counts().plot(kind="bar", color=["dodgerblue", "crimson", "orange", "darkgreen"], edgecolor="black")
plt.title("🎯 Squad Tactical Action Categories", fontsize=13, fontweight="bold")
plt.ylabel("Count of Players")
plt.xticks(rotation=0)
plt.show()
```

### 📈 Dashboard 8: Goals + Assists (Grouped Bar Chart)
**Question:** Who contributes most to the attack?

```python
x = np.arange(len(df["name"]))
width = 0.35

plt.figure(figsize=(12, 5))
plt.bar(x - width/2, df["total_goals"], width, label='Goals', color='blue', edgecolor='black')
plt.bar(x + width/2, df["total_assists"], width, label='Assists', color='cyan', edgecolor='black')

plt.title("⚔️ Attack Output Contribution Map", fontsize=13, fontweight="bold")
plt.xlabel("Players")
plt.ylabel("Count")
plt.xticks(x, df["name"], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

### 📈 Dashboard 9: Team Radar Comparison (Radar Chart)
**Question:** How does the average team balance compare across parameters?

```python
categories = ['Passing', 'Defending', 'Stamina', 'Speed', 'Shooting']
avg_values = [df['passing'].mean(), df['defending'].mean(), df['stamina'].mean(), df['speed'].mean(), df['shooting'].mean()]

angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
avg_values += avg_values[:1]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, avg_values, color='red', alpha=0.25)
ax.plot(angles, avg_values, color='red', linewidth=2)

ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, fontsize=11, fontweight="bold")
plt.title("🕸️ Team Average Skill Radar Profile", fontsize=14, fontweight="bold", y=1.1)
plt.show()
```

### 📈 Dashboard 10: Team Scoreboard
**Question:** Can we print an instant dashboard layout summarizing the entire club status?

```python
# Calculate scoreboards metrics
total_players = len(df)
avg_age = round(df["age"].mean(), 1)
avg_stamina = round(df["stamina"].mean(), 1)
need_rest_count = int(df["needs_rest"].sum())
need_shooting_count = int((df["shooting"] < 80).sum())
ready_count = int((df["status"] == "Match Ready").sum())
best_passer = df.loc[df["passing_accuracy"].idxmax(), "name"]

print("================================================")
print("                 ⚽ ARTIN FC ⚽                  ")
print("================================================")
print(f"👥 Active Players:           {total_players}")
print(f"📅 Average Age:              {avg_age} Years Old")
print(f"🫁 Average Stamina:          {avg_stamina}%")
print(f"💤 Needs Rest Recovery:      {need_rest_count} Players")
print(f"🎯 Needs Shooting Practice:  {need_shooting_count} Players")
print(f"🏃 Ready For Match Action:   {ready_count} Players")
print(f"🌟 Best Passer (Precision): {best_passer} ({df['passing_accuracy'].max()}%)")
print("================================================")
```

## 👑 Part 7: The Master Boss Challenge ⭐⭐⭐⭐⭐

Your final boss challenge is to write a single master function called `show_team_dashboard(players)`. 

When this function is run, it should completely build the entire analytics suite, print statistics, highlight key players, and show all analysis graphs automatically!

```python
def show_team_dashboard(players):
    """
    MASTER ANALYTICS APP
    Run this function to process and display the entire Artin FC report!
    """
    # 1. Build the DataFrame
    local_df = pd.DataFrame(players)
    
    # 2. Compute calculated metrics
    local_df['total_goals'] = local_df['goals'].apply(sum)
    local_df['total_assists'] = local_df['assists'].apply(sum)
    local_df["total_completed"] = local_df["passes_completed"].apply(sum)
    local_df["total_attempts"] = local_df["total_passes"].apply(sum)
    local_df["passing_accuracy"] = round((local_df["total_completed"] / local_df["total_attempts"]) * 100, 1)
    local_df["needs_rest"] = local_df["stamina"] < 82
    
    # Solve status mapping
    def map_local_status(p):
        if p["needs_rest"]:
            return "Recovery"
        elif p["shooting"] < 80:
            return "Shooting Practice"
        else:
            return "Match Ready"
    local_df["status"] = local_df.apply(map_local_status, axis=1)
    
    # 3. Print Team Scoreboard
    print("================================================")
    print("         🏁 ARTIN FC MASTER DASHBOARD 🏁         ")
    print("================================================")
    print(f"👥 Squad Size:            {len(local_df)} Players")
    print(f"📅 Average Age:            {round(local_df['age'].mean(), 1)} Years")
    print(f"🫁 Team Stamina Average:   {round(local_df['stamina'].mean(), 1)}%")
    print(f"💤 Players Needing Rest:   {int(local_df['needs_rest'].sum())}")
    print(f"🎯 Need Shot Training:    {int((local_df['shooting'] < 80).sum())}")
    print(f"⚽ Top Goal Scorer:        {local_df.loc[local_df['total_goals'].idxmax(), 'name']}")
    print(f"🌟 MVP (Overall Rating):   {local_df.loc[(local_df['passing']+local_df['stamina']+local_df['speed'])/3.0].idxmax(), 'name']}")
    print("================================================\n")
    
    # 4. Generate visual layout subplots
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    
    # Subplot 1: Goals bar
    axes[0, 0].bar(local_df["name"], local_df["total_goals"], color="forestgreen", edgecolor="black")
    axes[0, 0].set_title("Goals Scored")
    axes[0, 0].tick_params(axis='x', rotation=45)
    
    # Subplot 2: Assists bar
    axes[0, 1].bar(local_df["name"], local_df["total_assists"], color="gold", edgecolor="black")
    axes[0, 1].set_title("Assists Created")
    axes[0, 1].tick_params(axis='x', rotation=45)
    
    # Subplot 3: Passing accuracy horizontal rating
    sorted_sub = local_df.sort_values(by="passing_accuracy")
    axes[1, 0].barh(sorted_sub["name"], sorted_sub["passing_accuracy"], color="teal", edgecolor="black")
    axes[1, 0].set_title("Passing Accuracy (%)")
    
    # Subplot 4: Position distributions
    local_df["position"].value_counts().plot(kind="pie", autopct="%1.1f%%", ax=axes[1, 1], colors=["skyblue", "orange", "lightgreen", "violet"])
    axes[1, 1].set_ylabel("")
    axes[1, 1].set_title("Squad Position Balance")
    
    plt.tight_layout()
    plt.show()

# Let's run our master app with ONE line of code!
show_team_dashboard(Artin_FC_v2)
```
