## 🛠️ Part 3. Building Your First Chart Step-by-Step

###  Introduction to Matplotlib and Seaborn Libraries

**Matplotlib** is Python's canvas.
1. Declare a figure frame (`plt.figure()`).
2. Draw elements (bars, lines, scatters).
3. Customize layout (`plt.xlabel()`, `plt.title()`, `plt.grid()`).
4. Show it (`plt.show()`).

**Seaborn** sits on top of Matplotlib, automatically applying crisp color palettes like `viridis`, `flare`, or `crest`.


### 💡 Anatomy of a Plot in Python:
- `plt.figure(figsize=(width, height))`: Tells Python how wide and tall to build our blank canvas.
- `sns.barplot(x=..., y=..., data=df)`: Maps columns directly to the $X$ and $Y$ coordinates.
- `plt.title()`: Adds a header title.
- `plt.xlabel()` and `plt.ylabel()`: Label our chart borders.
- `plt.xticks(rotation=45)`: Tilts our player names sideways so they don't overlap!
- `plt.show()`: Tells Python to render the final image on our screen.

Let's see how each line of code acts as a brush stroke.

```python
# 1. Initialize Canvas
plt.figure(figsize=(10, 5))

# 2. Draw Bars
sns.barplot(x="name", y="speed", data=df)

# 3. Label Canvas
plt.title("⚡ Squad Speed Index (Descending)", fontsize=14, fontweight="bold")
plt.xlabel("Player Roster", fontsize=11)
plt.ylabel("Speed Rating", fontsize=11)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 🔄 The Data Science Secret: Wide vs. Long Data

Before we look at our final dashboard, we need to learn a quick trick that professional data engineers use.

Up until now, our team data has been in a **Wide Format**. It looks like a standard spreadsheet where every skill has its own column:

| name | passing | defending | shooting |
| :--- | :---: | :---: | :---: |
| **Artin** | 85 | 45 | 88 |

This is great for humans to read, but advanced graphing tools like Seaborn struggle with it when we want to compare multiple skills on the *same* chart. Seaborn prefers **Long Format**, where data is stacked into a single column.

To fix this, we use a powerful Pandas tool called `pd.melt()`. It acts like a zipper, unpivoting the columns so they stack vertically:

| name | Attribute (var_name) | Rating (value_name) |
| :--- | :---: | :---: |
| **Artin** | passing | 85 |
| **Artin** | defending | 45 |
| **Artin** | shooting | 88 |

Now that all the skills are neatly listed under one column (`Attribute`) and all the scores are under another (`Rating`), Seaborn can effortlessly draw our final team signature plot!

### 🏟️ Dashboard 10 – Team Performance Summary
**Tactical Goal:** Construct a single multi-panel chart comparing three critical technical ratings: passing, defending, and shooting side-by-side using violin distributions.

```python
melted_skills = pd.melt(df, id_vars=['name'], value_vars=['passing', 'defending', 'shooting'],
                        var_name='Attribute', value_name='Rating')

plt.figure(figsize=(10, 6))
sns.violinplot(x="Attribute", y="Rating", data=melted_skills, inner="point", palette="coolwarm")
plt.title("🏟️ Team Core Skills Signature Distributions", fontsize=14, fontweight="bold")
plt.xlabel("Tactical Dimension")
plt.ylabel("Attribute Rating Scale (0-100)")
plt.tight_layout()
plt.show()
```

## 🛠️ Part 5. Building the Complete Team Dashboard Function
Let's build the interactive command console that brings it all together.

```python
def show_team_dashboard(dataframe):
    # ---------------------------------------------------------
    # STEP 1: PRINT THE HEADER TEXT FOR THE CONSOLE
    # ---------------------------------------------------------
    # We print a clean visual header using string multiplication "=" * 60
    print("=" * 60)
    print("⚽ ARTIN FC SYSTEM SCOUTING CONSOLE ⚽")
    print("=" * 60)

    # ---------------------------------------------------------
    # STEP 2: CALCULATE THE SQUAD STATISTICS (KPIs)
    # ---------------------------------------------------------
    # 1. Sum up all the goals scored by every player in the squad
    total_squad_goals = dataframe['total_goals'].sum()

    # 2. Calculate the average passing accuracy across the entire team
    avg_passing_accuracy = dataframe['passing_accuracy'].mean()

    # 3. Find the name of the player who has the highest combined goals + assists
    # idxmax() finds the index of the highest value, and .loc gets their 'name'
    top_contributor = dataframe.loc[dataframe['goal_contributions'].idxmax(), 'name']

    # ---------------------------------------------------------
    # STEP 3: DISPLAY THE STATS IN THE CONSOLE
    # ---------------------------------------------------------
    # We use f-strings to inject our calculated numbers directly into the text.
    # :.2f limits the passing accuracy percentage to just 2 decimal places.
    print(f"🔹 Total Seasonal Team Goals: {total_squad_goals}")
    print(f"🔹 Average Squad Passing Efficiency: {avg_passing_accuracy:.2f}%")
    print(f"🔹 Star Offensive Playmaker: {top_contributor}")
    print("-" * 60)

    # ---------------------------------------------------------
    # STEP 4: SET UP THE MULTI-CHART CANVAS (SUBPLOTS)
    # ---------------------------------------------------------
    # plt.subplots(1, 2) splits our visual window into 1 row and 2 columns (side-by-side).
    # 'fig' controls the whole window frame, and 'axes' acts like an array containing our 2 plots:
    # axes[0] is the left plot, axes[1] is the right plot.
    # figsize=(16, 6) makes the window 16 inches wide and 6 inches tall.
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    # ---------------------------------------------------------
    # STEP 5: BUILD PANEL A - THE LEFT CHART (BARPLOT)
    # ---------------------------------------------------------
    # 1. Sort the players by their goal contributions so the highest is first
    sorted_df = dataframe.sort_values("goal_contributions", ascending=False)

    # 2. Draw the barplot. Notice ax=axes[0] -> this forces Seaborn to draw it on the LEFT chart!
    sns.barplot(ax=axes[0], x="name", y="goal_contributions", data=sorted_df, palette="crest")

    # 3. Set the title for the left chart specifically using axes[0]
    axes[0].set_title("Direct Goal Contributions (Goals + Assists)", fontsize=12, fontweight="bold")

    # 4. Rotate the player names on the X-axis by 45 degrees so they don't overlap each other
    axes[0].tick_params(axis='x', rotation=45)

    # ---------------------------------------------------------
    # STEP 6: BUILD PANEL B - THE RIGHT CHART (SCATTER PLOT)
    # ---------------------------------------------------------
    # 1. Draw a scatter plot. Notice ax=axes[1] -> this forces Seaborn to draw it on the RIGHT chart!
    # x="passing" and y="defending" maps their tactical profile on the grid.
    # hue="position" colors the dots by their role (Forward, Defender, etc.).
    # s=180 makes the scatter plot dots extra large and easy to see.
    sns.scatterplot(ax=axes[1], x="passing", y="defending", hue="position", s=180, data=dataframe)

    # 2. Set the title for the right chart specifically using axes[1]
    axes[1].set_title("Positioning Map: Defense vs. Passing Abilities", fontsize=12, fontweight="bold")

    # ---------------------------------------------------------
    # STEP 7: CLEAN UP LAYOUT AND SHOW THE FINAL DASHBOARD
    # ---------------------------------------------------------
    # tight_layout() automatically adjusts spacing so nothing gets cut off or squeezed
    plt.tight_layout()

    # Finally, render the beautiful combined dashboard on the screen!
    plt.show()

# Run the function using our team DataFrame to see the console and charts!
show_team_dashboard(df)
```
