The head coach has just requested a comprehensive **Pre-Match Analytics Dossier** for today's tournament fixture.

```python
# 🛠️ TASK 1 & 2: Import your libraries and load the player data matrix
import coach_toolbox_v3 as ct
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert Artin_FC_v2 dataset into a Pandas DataFrame named 'df'
df = pd.DataFrame(Artin_FC_v2)
df
```

```python
def elite_status(player):
    player_rating = ct.calculate_overall_rating(player)
    if player_rating > 85:
        print("ELITE status")
    else:
        print("NOT elite status")

elite_status(Artin_FC_v2[0])
```

```python
import coach_toolbox_v3 as ct
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import importlib
importlib.reload(ct)


# 1. Create a column 'goal_contributions' that adds goals + assists together
df['total_goals'] = df['goals'].apply(sum)
df['total_assists'] = df['assists'].apply(sum)
df['goal_contributions'] = df['total_goals'] + df['total_assists']

# Calculate player rating first, as 'elite_status' depends on it
df['rating'] = df.apply(ct.calculate_overall_rating, axis=1)

# 2. Create a column 'needs_rest'
# Hint: If stamina < 75, then "Rest Needed", else "Match Ready"
df['needs_rest'] = df.apply(ct.needs_rest, axis=1)

# 3. Create a column 'elite_status' indicating players with a performance rating >= 8.5
# Hint: Use df['rating'] >= 8.5 mapping
df['ELITE_STATUS'] = df['rating'] >= 80

df
```

```python
# 🛠️ TASK 4: Tactical Intelligence Q&A
# Write small code queries to fetch the exact answers for the coach:

# Q1: Which individual player registered the absolute most goals?
top_scorer_identity = df["total_goals"].idxmax()
top_scorer_goals = df["total_goals"].max()
top_scorer_name = df.loc[top_scorer_identity, "name"]
print(f"⚽ Top Goal Scorer: {top_scorer_name} with {top_scorer_goals} goals.")

# Q2: Which specific Midfielder has the highest stamina score?
midfielders = df[df["position"] == "Midfielder"]
fit_midfielder_row = midfielders["stamina"].idxmax()
fit_midfielder_identity = midfielders.loc[fit_midfielder_row, "name"]
print(f"{fit_midfielder_identity} is the fittest midfielder.")

# Q3: What is the average calculated performance rating of the entire team?
print("📊 Team Average Rating:", df["rating"].mean())

# Q4: Filter and display only the rows of players who currently show 'Rest Needed':
print("⚠️ Squad Members Requiring Rest:")
# The 'needs_rest' column was already created in TASK 3.
# We filter the DataFrame 'df' using the boolean values in 'needs_rest'
# and then select the 'name' column.
print(df[df["needs_rest"]][["name"]])
```

```python
# 🛠️ TASK 5: Visual Scouting Suite
# Using your freedom of layout design, create 3 distinct visual charts.
# Remember to add titles, labels, and set clear styles!

# Chart A: Visual comparison of direct Goals Scored by player
plt.figure(figsize=(6,4))
# 2. Draw Bars
sns.barplot(x="name", y="total_goals", data=df)

# 3. Label Canvas
plt.title("GOALS SCORED BY PLAYER)", fontsize=14, fontweight="bold")
plt.xlabel("Player Roster", fontsize=11)
plt.ylabel("GOALS", fontsize=11)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Chart B: Stamina Spectrum across the squad roster
plt.figure(figsize=(6,4))
sns.barplot(x="name", y="stamina", data=df)

# 3. Label Canvas
plt.title("STAMINA BY PLAYER)", fontsize=14, fontweight="bold")
plt.xlabel("Player Roster", fontsize=11)
plt.ylabel("STAMINA", fontsize=11)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Chart C: Relationship Mapping between Stamina (X) and Match Rating (Y)
plt.figure(figsize=(6,4))
sns.scatterplot(x="stamina", y="rating", hue="position", style="position", s=180, data=df)
plt.title("Player Stamina and Player Rating", fontsize=14, fontweight="bold")
plt.xlabel("Stamina")
plt.ylabel("Match Rating")
plt.tight_layout()
plt.show()
```
