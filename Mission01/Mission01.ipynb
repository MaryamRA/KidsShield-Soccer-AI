## ⚽ Part 8— Match Report Generator

```python
home_team = "France"
away_team = "Germany"

home_goals = 3
away_goals = 1

print("⚽ Final Match Report ⚽")
print("-----------------------")
print(home_team, home_goals, "-", away_goals, away_team)

if home_goals > away_goals:
    print("Winner:", home_team)
elif away_goals > home_goals:
    print("Winner:", away_team)
else:
    print("Result: Draw")
```
