#### 🧩 Challenge 2.1: The Custom Overall Rating Calculator
**Goal:** Let's define an updated overall rating function that calculates a player's total performance.
For outfield players (everyone except the goalkeeper 'GK'), their overall score is the mathematical average of all 4 skill attributes.

For Goalkeepers ('GK'), defending and shooting aren't as important, so their overall rating is calculated differently: ` (stamina * 0.40) + (speed * 0.20)`.

````python

def calculate_overall_rating(player):
    # your code here
    if player["position"] == "GK":
        rating  = (player["stamina"] * 0.40) + (player["speed"] * 0.20)
    else:
        rating = (player["defending"] + player["stamina"] + player["speed"] + player["shooting"]) / 4

    return rating


print(calculate_overall_rating(Artin_FC_v2[-1]))

````

### 🧩 Challenge 3.1: The Tactical AI Agent recommendation Engine
Let's write a function that reviews a player dictionary and returns a tailored coaching card:
- If a player's defending attribute is greater than 80, recommend: `"Solid defensive anchor. Hold the backline."`
- If their speed is greater than 90 and passing is greater than 80, recommend: `"Winger breakout threat. Focus on flank overlaps."`
- If their shooting is greater than 90, recommend: `"Target striker. Look for direct shots inside the box."`
- Otherwise, return: `"Tactical supporting playmaker. Distribute cleanly."`

````python
def generate_ai_tactical_recommendation(player):
    # your code here
    if player["defending"] > 80:
        return "Solid defensive anchor. Hold the backline."

    if player["speed"] > 90 and player["shooting"] > 80:
        return "Winger breakout threat. Focus on flank overlaps."

    elif player["shooting"] > 90:
        return "Target striker. Look for direct shots inside the box."

    else:
        return "Tactical supporting playmaker. Distribute cleanly."
````

````python
from coach_toolbox_v2 import needs_rest

def analysis_report_v1(team):

    total_players = len(team)

    need_rest = 0

    for player in team:

        if needs_rest(player):
            need_rest += 1


    report = {
        "Number of players": total_players,
        "Number needing rest": need_rest
    }

    return report

print(analysis_report_v1(Artin_FC_v2))
````

## 🏆 Bonus Challenge
Create a function named `analysis_report_v2(team)` that returns:

````python
# Write your solution here
from coach_toolbox_v2 import *

def analysis_report_v2(team):

    total_players = len(team)

    need_rest = 0

    for player in team:

        if needs_rest(player):
            need_rest += 1

    needs_passing = 0

    for player in team:

        if needs_passing_practice(player):
            needs_passing += 1

    ready = 0

    for player in team:
        if coach_decision(player) == "Ready for Match.":
            ready += 1




    report = {
        "Number of players": total_players,
        "Number needing rest": need_rest,
        "Number needing passing practice": needs_passing,
        "Number ready for match": ready
    }

    return report
````
