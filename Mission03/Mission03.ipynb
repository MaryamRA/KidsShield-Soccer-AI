## 🟩 Part 14 — while Loops
A football match continues running **while** time remains. Unlike `for` loops that step through a collection, `while` loops run continuously as long as a statement stays True.

````python
minute = 1

while minute <= 10:
    print(minute)
    minute += 1
````

## 🟩 Part 15 — 90 Minute Match Clock
Now the fun begins! Let's build a match clock engine.

````python
minute = 1

while minute <= 90:
    print(f"Minute {minute}")
    minute += 1
````

## 🟩 Part 16 — Special Match Events
We can catch specific timeline moments by putting an `if` statement right inside our running clock loop.

````python
minute = 1

while minute <= 90:
    if minute == 45:
        print("Half Time")
    if minute == 90:
        print("Full Time")
    minute += 1
````

## 🟩 Part 17 — Bringing the Game to Life with Randomness! 🎲⚽
Right now, our match clocks are completely static. But real football is unpredictable! 

To fix this, we import Python's `random` module. Every single minute, the computer rolls a 100-sided mathematical die. If the die lands on a high number (e.g., above 98), a goal is generated out of thin air!

````python
import random

minute = 1
home_score = 0
away_score = 0

print("🎬 MATCH STARTING...")

while minute <= 90:
    # Roll a 100-sided die for the Home Team attack
    home_dice = random.randint(1, 100)
    # Roll a 100-sided die for the Away Team attack
    away_dice = random.randint(1, 100)
    
    # 2% chance to score each minu
    if home_dice > 98:
        home_score += 1
        print(f"⚽ GOAL HOME TEAM! Minute {minute} | Score: {home_score}-{away_score}")
        
    if away_dice > 98:
        away_score += 1
        print(f"⚽ GOAL AWAY TEAM! Minute {minute} | Score: {home_score}-{away_score}")
        
    # Check for half-time structural whistle
    if minute == 45:
        print(f"🧡 HALF TIME Whistle! Current Score: {home_score}-{away_score}")
        
    minute += 1

print("🏁 FULL TIME!")te
print(f"Final Scoreboard: Home {home_score} - {away_score} Away")
````

````python
# This is a simpler version of the above code

import random

team_a_score = 0
team_b_score = 0

print("🎬 MATCH STARTING...")

for minute in range(1, 91):
    # Team A attack
    if random.randint(1, 100) > 97:
        team_a_score += 1
        print("🔥 Team A scored at minute", minute)
        
    # Team B attack
    if random.randint(1, 100) > 97:
        team_b_score += 1
        print("🔥 Team B scored at minute", minute)
````
