You successfully solved your bonus challenge; an analysis_report-v2(team) function. Let's review and test it again.

```python
from coach_toolbox_v2 import *

def analysis_report_v2(team):

    total_players = len(team)
    need_rest = 0
    needs_passing = 0
    ready = 0

    for player in team:
        if needs_rest(player):
            need_rest += 1

    for player in team:
        if needs_passing_practice(player):
            needs_passing += 1

    for player in team:
        if coach_decision(player) == "Ready for Match.":
            ready += 1

    report = {
        "Number of players": total_players,
        "Number needing rest": need_rest,
        "Number needing passing practice": needs_passing,
        "Number ready for match": ready
    }

    return report
```

```python
team_report = analysis_report_v2(Artin_FC_v2)
print(team_report)
```

* **Rows:** Each row (horizontal) represents a single **individual player**.
* **Columns:** Each column (vertical) represents a specific **attribute or metric** of information about that player.



### 🗺️ Introduction: How Does .loc[] Work?

### Hunting Down the Top Scorer (Step-by-Step)
we are going to learn how to load your custom function from yesterday's library, map it onto our DataFrame spreadsheet, find the highest value, and introduce the ultimate data mapping tool: .loc (Location Selection).

#### 📦 Step 1: Import the Custom Function from your Library
Before Pandas can do any work, we need to fetch our specialized calculation tool from the external coach_toolbox.py library file.

```python
from coach_toolbox_v2 import calculate_goals

print(" goal calculator successfully imported from coach_toolbox!")
```

#### 🧬 Step 2: Use .apply() to Create the New Column
Now, we want to run your function across every single row of our DataFrame. Instead of writing a slow for loop, Pandas has a superpower called .apply().

We pass our custom function inside .apply() and tell it axis=1 (which means "apply this across every horizontal row").

```python
# Create a brand new column named 'total_goals' by passing the player rows through your function
df['total_goals'] = df.apply(calculate_goals, axis=1)

# Let's peek at our updated team grid!
df[['name', 'position', 'total_goals']]
```

*Look at that! Pandas automatically looped through the team and added a clean total number right next to each player's name.*

#### 🔢 Step 3: Find the Row Label Using idxmax()
Now that we have a column of numbers, we use idxmax() (which stands for "Index of the Maximum").

idxmax() scans down the total_goals column, finds the largest number, and returns the Row Index Number where that record lives!

```python
# Ask Pandas for the exact row index label where the highest value lives
highest_row_index = df['total_goals'].idxmax()

print(f"🎯 The top scorer is located at row index number: {highest_row_index}")
```

#### 👑 Step 4: Use .loc[] to Print the Winner's Card
Now we combine everything! We hand our newly discovered best_row_index straight to .loc[] to instantly extract their identity and final score:

```python
df.loc[highest_row_index, "name"]
```

```python
# Pass the row index into .loc[row, column] to extract the data cards
top_scorer_name = df.loc [highest_row_index, 'name']
highest_goals = df.loc[highest_row_index, 'total_goals']


# Print the final official scoreboard report!
print("==================================================")
print(f"🏆 GOLDEN BOOT WINNER")
print("==================================================")
print(f"Top Scorer Identity: {top_scorer_name}")
print(f"Total Output Score:  {highest_goals} Goals")
print("==================================================")
```

```python
# Combined code for hunting the top scorer

# Step 1: Import the custom tool from your library
from coach_toolbox_v2 import calculate_goals

# Step 2: Apply across rows to create our total goals column
df['total_goals'] = df.apply(calculate_goals, axis=1)

# Step 3: Find the exact row index of the highest value using idxmax()
top_player_row = df['total_goals'].idxmax()

# Step 4: Use .loc[row, column] to pull the name and score instantly
top_scorer_identity = df.loc[top_player_row, 'name']
highest_score = df.loc[top_player_row, 'total_goals']

# Display our official analytics report!
print(f"⚽ Top Goal Scorer: {top_scorer_identity} with {highest_score} total goals!")
```

## 🧬 Part 5: Add New Columns

We can expand our active DataFrame dynamically! Let's compute two major columns:
1. **Passing Accuracy:** The ratio of completed passes over total attempts.
2. **Needs Rest:** A boolean column that checks if their average stamina is falling low.

```python
# 1. Compute total completed passes over total passes
df["total_completed"] = df["passes_completed"].apply(sum)
df["total_attempts"] = df["total_passes"].apply(sum)
df["passing_accuracy"] = round((df["total_completed"] / df["total_attempts"]) * 100, 1)

# 2. Calculate if they need rest (if their active stamina rating is less than 82)
df["needs_rest"] = df["stamina"] < 82

# Let's view our updated DataFrame columns!
df[["name", "position", "passing_accuracy", "stamina", "needs_rest"]]
```
